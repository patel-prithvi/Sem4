# where function

```python
import numpy as np
df= pd.DataFrame({'Marks':[40,35,65,31,70,95,18]})
df
```

```python
df['Result']=np.where(df['Marks']>40, 'Pass', 'False')
df
```

```python
df['Grade']=np.where(df['Marks']>75, 'A', np.where(df['Marks']>50, 'B', 'C'))
df
```

# Train_test_split

```python
from sklearn.model_selection import train_test_split
df=pd.read_csv('car data.csv')
df.drop(columns=['Car_Name'], inplace=True)
df= pd.get_dummies(df,drop_first=True)
y= df['Selling_Price']
x= df.drop(columns=['Selling_Price'])
```

```python
print("x=", x.shape)
print("y=", y.shape)
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.2, random_state=42)
```
