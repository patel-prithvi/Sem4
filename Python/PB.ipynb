```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.DataFrame({
    "Math":[78,85,90,66,88],
    "Science":[80,82,89,70,85],
    "English":[75,88,92,60,84],
    "Computer":[90,91,95,72,89]
}, index=["S1","S2","S3","S4","S5"])
sns.heatmap(df,annot=True,cmap="coolwarm", linewidths=1, linecolor="black")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.DataFrame({
    "Height":[150,160,170,180,175],
    "Weight":[50,60,65,75,70],
    "Age":[20,22,25,30,28],
    "BMI":[22,24,23,26,25]
})
sns.heatmap(df.corr(),cmap="YlGnBu")
plt.show()
```
