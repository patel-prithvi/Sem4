# Pandas Transformation function

```python
import pandas as pd 
data = {'A':[10,20,30], 'B':[5,15,25]}
df=pd.DataFrame(data)
df
```

```python
# apply()
df.apply(sum)
```

```python
df.apply(sum, axis=1)
```

# using lambda function

```python
df.apply(lambda x:x**2)
```

# using user define function

```python
def square(x):
    return x*x
def cube(x):
    return x*x*x
```

```python
df["square"]=df['A'].apply(square)
df["cube"]=df['B'].apply(cube)
df
```

# using function parameter

```python
def calc(x,a,b):
    return x*(a+b)
df['result']= df['A'].apply(calc, args=(3,4))
df
```

# astype()

```python
df['result']=df['result'].astype(float)
df
```

# replace()

```python
df.replace(10,1001)
```

```python
df.replace({5:50, 15:155, 25:255})
```

# groupby()

```python
data={"Department":['IT','HR','IT','HR','Sales','IT'],
     "Employee":['A','B','C','D','E','F'],
     'Salary':[50000,45000,60000,47000,52000,70000]}
df=pd.DataFrame(data)
df
```

```python
df.groupby('Department')['Salary'].sum()
```

```python
df.groupby('Department')['Salary'].mean()
```

```python
df.groupby('Department')['Salary'].median()
```

# multiple aggregation

```python
df.groupby('Department')['Salary'].agg(['sum', 'mean', 'max'])
```

# apply() for custom operation

```python
df.groupby('Department')['Salary'].apply(lambda x:x.max()-x.min())
```

# concat()

```python
df1=pd.DataFrame({'A':[1,2],'B':[3,4]})
df2=pd.DataFrame({'A':[5,6],'B':[7,8]})
df1
```

```python
df2
```

```python
result = pd.concat([df1,df2])
result
```

```python
result = pd.concat([df1,df2], axis=1)
result
```

```python
df1=pd.DataFrame({'A':[1,2],'B':[3,4]})
df2=pd.DataFrame({'A':[5,6],'D':[7,8]})
result=pd.concat([df1,df2],join='inner') 
result
```

```python
result=pd.concat([df1,df2],join='outer')
result
```

# ignore_index()

```python
df1=pd.DataFrame({'A':[1,2],'B':[3,4]})
df2=pd.DataFrame({'A':[5,6],'D':[7,8]})
result=pd.concat([df1,df2],ignore_index='true') 
result
```

# keys

```python
df1=pd.DataFrame({'A':[1,2],'B':[3,4]})
df2=pd.DataFrame({'A':[5,6],'B':[7,8]})
result=pd.concat([df1,df2],keys=["train", "test"]) 
result
```

# names

```python
df1=pd.DataFrame({'A':[1,2],'B':[3,4]})
df2=pd.DataFrame({'A':[5,6],'B':[7,8]})
result=pd.concat([df1,df2],keys=["train", "test"],names=["Dataset"]) 
result
```

# verify_integrity(default=False)

```python
df1=pd.DataFrame({'A':[1,2],'B':[3,4]})
df2=pd.DataFrame({'A':[5,6],'B':[7,8]})
result=pd.concat([df1,df2],verify_integrity=False) 
result
```

# sort()

```python
df1=pd.DataFrame({'E':[1,2],'B':[3,4]})
df2=pd.DataFrame({'D':[5,6],'C':[7,8]})
result=pd.concat([df1,df2],sort=True) 
result
```

```python
# 13
Q1 = df['popularity'].quantile(0.25)
Q3 = df['popularity'].quantile(0.75)
IQR = Q3 - Q1

outliers_popularity = df[(df['popularity'] < Q1 - 1.5*IQR) | (df['popularity'] > Q3 + 1.5*IQR)]

print("\nOutliers in popularity column:")
outliers_popularity
```
