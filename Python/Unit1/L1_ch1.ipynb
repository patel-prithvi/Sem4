# Pandas Series

```python
# create series by using list
import pandas as pd
marks= [85,90,70,65]
s= pd.Series(marks)
print(s)
```

```python
# Create series by using customise serial number
import pandas as pd
marks= [85,90,70,65]
s= pd.Series(marks, index= ['DM', 'FSD', 'COA', "TOC"])
print(s)
```

```python
# Create series by using tuple
import pandas as pd
t= (85,90,70,65)
s= pd.Series(t)
print(s)
```

```python
# Create series by using dict
import pandas as pd
d= {'a':10, 'b':20, 'c':30}
s= pd.Series(d)
print(s)
```

```python
# create series by using array
import numpy as np
import pandas as pd
marks= np.array([85,90,70,65])
s= pd.Series(marks)
print(s)
```

```python
# create series by using range fun
import pandas as pd
s= pd.Series(range(11,18))
print(s)
```


# DataFrame

```python
# create dataframe by using dict
import pandas as pd
data= {'name':['a','b','c'], 'marks':[100,99,98], 'City':['ahm','vad','gandhinagar']}
df=pd.DataFrame(data,index=[1,2,3])
df
```

```python
# Create dataframe by using list
import pandas as pd
data= [['a',80],['b',90],['c',100]]
df=pd.DataFrame(data,columns=['names','marks'],index=[1,2,3])
df
```

```python
# create dataframe by using pandas series
import pandas as pd
s1= pd.Series([1,2,3])
s2= pd.Series([8,9,10])
l={'col1':s1,'col2':s2}
df=pd.DataFrame(l)
df
```
