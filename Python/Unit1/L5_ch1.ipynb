# merge Dataframe

```python
import pandas as pd 
employee = pd.DataFrame({'emp_id':[1,2,3],
                        'name':['A','B','C'],
                        'dept_id':[10,20,30]})
department = pd.DataFrame({'dept_id':[10,20,40],
                          'dept_name':['HR','IT','Finance']})
pd.merge(employee, department, on='dept_id', how='inner')
```

```python
pd.merge(employee, department, on='dept_id', how='outer')
```

```python
pd.merge(employee, department, on='dept_id', how='left')
```

```python
pd.merge(employee, department, on='dept_id', how='right')
```

# left_on & right_on

```python
employee = pd.DataFrame({'emp_id':[1,2],
                        'name':['A','B'],
                        'dept_id':[10,20]})
department = pd.DataFrame({'d_id':[10,20],
                          'dept_name':['HR','IT']})
pd.merge(employee, department, left_on="dept_id", right_on="d_id")
```

# left_index & right_index

```python
pd.merge(employee, department, left_index=True, right_index=True)
```

# suffixes

```python
employee = pd.DataFrame({'emp_id':[1,2],
                        'name':['A','B'],
                        'dept_id':[10,20]})
department = pd.DataFrame({'dept_id':[10,20],
                          'dept_name':['HR','IT'],
                          'name':['X','Y']})
pd.merge(employee, department, on="dept_id", suffixes=('_emp','_dept'))
```

# indicator

```python
employee = pd.DataFrame({'emp_id':[1,2,3],
                        'name':['A','B','C'],
                        'dept_id':[10,20,30]})
department = pd.DataFrame({'dept_id':[10,20,40],
                          'dept_name':['HR','IT','Finance']})
pd.merge(employee, department, on='dept_id', how='outer', indicator= True)
```
