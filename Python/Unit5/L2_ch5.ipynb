# PB 268

```python
import pandas as pd
```

```python
df= pd.read_csv('data_breast_cancer.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df.drop(columns=['id','Unnamed: 32'],inplace=True)
df.head()
```

```python
df.columns
```

```python
y=df['diagnosis']
x=df[['radius_mean', 'texture_mean', 'perimeter_mean',
       'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
nn= KNeighborsClassifier(n_neighbors=21)
```

```python
model = nn.fit(x_train, y_train)
```

```python
y_pred= model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
TN=cm[0][0]
FP=cm[0][1]
FN=cm[1][0]
TP=cm[1][1]
```

```python
accuracy = (TP+TN)/(TP+TN+FP+FN)
error_rate = (FP+FN)/(TP+TN+FP+FN) 
sensitivity = TP/(TP+FN)
specificity = TN/(TN+FP)

print('accuracy =',accuracy)
print('error rate =',error_rate)
print('sensitivity =',sensitivity)
print('specificity =',specificity)
```

```python
df1= pd.read_csv('PlayTennis.csv')
df1.head()
```

```python
df1.isna().sum()
```

```python
df1= pd.get_dummies(data=df1, drop_first=True)
df1.head()
```

```python
y1=df1['play_yes']
x1=df1[['windy', 'outlook_rainy', 'outlook_sunny', 'temp_hot', 'temp_mild', 'humidity_normal']]
```

```python
from sklearn.model_selection import train_test_split
x_train1, x_test1, y_train1, y_test1 = train_test_split(x1,y1,test_size=0.2,random_state=42)
```

```python
print(x_train1.shape)
print(x_test1.shape)
print(y_train1.shape)
print(y_test1.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
nn= KNeighborsClassifier(n_neighbors=3)
```

```python
model1 = nn.fit(x_train1, y_train1)
```

```python
y_pred1= model1.predict(x_test1)
y_pred1
```

```python
from sklearn.metrics import confusion_matrix
cm1 = confusion_matrix(y_test1,y_pred1)
cm1
```

```python
TN=cm1[0][0]
FP=cm1[0][1]
FN=cm1[1][0]
TP=cm1[1][1]
```

```python
accuracy = (TP+TN)/(TP+TN+FP+FN)
error_rate = (FP+FN)/(TP+TN+FP+FN) 
sensitivity = TP/(TP+FN)
specificity = TN/(TN+FP)

print('accuracy =',accuracy)
print('error rate =',error_rate)
print('sensitivity =',sensitivity)
print('specificity =',specificity)
```

```python
# rainy, cool, high humidity, windy
model1.predict([[1,1,0,0,0,0]])
```
