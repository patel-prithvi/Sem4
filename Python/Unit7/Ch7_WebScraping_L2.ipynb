```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
```

```python
with open('30 Most Popular Movies.html', 'r', encoding='utf-8') as file:
    content = file.read()
soup = BeautifulSoup(content,'html.parser')
box=soup.find_all('div',class_="article_movie_title")
names=[]
rating=[]
for i in box:
    names.append(i.find('a').text.strip())
    rating.append(i.find('span', class_='tMeterScore').text.strip())
movies={'name':names, 'rating':rating}
df=pd.DataFrame(movies)
df
```

```python
# car name, mileage, price, rating
with open('Certified used Mercedes-Benz for sale.html', 'r', encoding='utf-8') as file:
    content = file.read()
soup = BeautifulSoup(content,'html.parser')
box=soup.find_all('div',class_="vehicle-details")
names=[]
mileage=[]
price=[]
rating=[]
for i in box:
    names.append(i.find('h2',class_='title').text.strip())
    mileage.append(i.find('div',class_='mileage').text.strip())
    price.append(i.find('span',class_='primary-price').text.strip())
    r=i.find('span', class_='sds-rating__count')
    if(r):
        rating.append(r.text.strip())
    else:
        rating.append(0)
cars={'name':names, 'mileage':mileage, 'price':price, 'rating':rating}
df=pd.DataFrame(cars)
df
```

```python
with open('flipkart.html', 'r', encoding='utf-8') as file:
    content = file.read()
soup = BeautifulSoup(content,'html.parser')
box=soup.find_all('div',class_="_3pLy-c row")
names=[]
rating=[]
a_price=[]
d_price=[]
diff=[]
for i in box:
    names.append(i.find('div',class_='_4rR01T').text.strip())
    rating.append(i.find('div',class_='_3LWZlK').text.strip())
    d=i.find('div',class_='_30jeq3 _1_WHN1').text.strip()
    d_price.append(d)
    a=i.find('div', class_='_3I9_wc _27UcVY')
    if(a):
        a_price.append(a.text.strip())
        dif= int(a.text.strip().replace('₹','').replace(',','')) - int(d.replace('₹','').replace(',',''))
        diff.append(dif)
    else:
        a_price.append(0)
        diff.append(int(d.replace('₹','').replace(',','')))
phones={'name':names, 'rating':rating, 'd_price':d_price, 'a_price':a_price, 'diff': diff}
df=pd.DataFrame(phones)
df
```

```python
# name of job, company, location
with open('indeed jobs.html', 'r', encoding='utf-8') as file:
    content = file.read()
soup = BeautifulSoup(content,'html.parser')
box=soup.find_all('td',class_="resultContent css-1o6lhys eu4oa1w0")
names=[]
company=[]
location=[]
for i in box:
    names.append(i.find('span').text.strip())
    company.append(i.find('span',class_='css-1h7lukg eu4oa1w0').text.strip())
    location.append(i.find('div',class_='css-1restlb eu4oa1w0').text.strip())
jobs={'name':names, 'company':company, 'location':location}
df=pd.DataFrame(jobs)
df
```
