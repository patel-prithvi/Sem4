# Multiple Linear Regression

```python
import pandas as pd
```

```python
# step 1 
df= pd.read_csv('car data.csv')
df.head()
```

```python
# step 2 check null value
df.isna().sum()
```

```python
#  step 3 data preprocessing
df['age']=2026-df['Year']
df.head()
```

```python
df.drop(columns='Year', inplace=True)
df.head()
```

```python
df.drop(columns='Car_Name', inplace=True)
df.head()
```

```python
df= pd.get_dummies(data=df, drop_first=True)
df.head()
```

```python
# step 4 define x & y
y= df['Selling_Price']
x=df[['Present_Price', 'Kms_Driven', 'Owner', 'age',
       'Fuel_Type_Diesel', 'Fuel_Type_Petrol', 'Seller_Type_Individual',
       'Transmission_Manual']]
```

```python
df.columns
```

```python
# step 5 split data into train & test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# step 6 create model
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
# step 7 find m & c
m=lr.coef_
m
```

```python
c= lr.intercept_
c
```

```python
# 8 , petrol, 45000, m, indi, 0, 900000
i20_price= lr.predict([[9,45000,0,8,0,1,1,1]])
i20_price
```

```python
# 'Present_Price', 'Kms_Driven', 'Owner', 'age', 'Fuel_Type_Diesel', 'Fuel_Type_Petrol', 'Seller_Type_Individual', 'Transmission_Manual'
# 9.54, 43000km , 0 owner, 13 age , diesel, manual, d
lr.predict([[9.54,43000,0,13,1,0,0,1]])
```

```python
# step 8 find MSE
y_predict=lr.predict(x_test)
y_predict
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
print("MSE=",mean_squared_error(y_test,y_predict))
print("MAE=",mean_absolute_error(y_test,y_predict))
print("R2_score=",r2_score(y_test,y_predict))
```
