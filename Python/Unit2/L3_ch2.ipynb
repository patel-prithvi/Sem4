# Plotly

# Bar chart

```python
import pandas as pd
import plotly.express as px
data={'Department':['IT','IT','HR','HR','Finance','Finance'],
     'Gender':['Male','Female','Male','Female','Male','Female'],
     'Salary':[50000,55000,45000,48000,70000,72000]}
df=pd.DataFrame(data)
fig= px.bar(df,x='Department', y='Salary', title='Department Salary ')
fig.show()
```

```python
fig= px.bar(df,x='Department', y='Salary', title='Department Salary', color="Gender", barmode='group')
fig.show()
```

```python
fig= px.bar(df,x='Department', y='Salary', title='Department Salary', color="Gender", barmode='stack')
fig.show()
```

```python
import pandas as pd
import plotly.express as px
data={'Department':['IT','IT','HR','HR','Finance','Finance'],
     'Gender':['Male','Female','Male','Female','Male','Female'],
     'Salary':[50000,55000,45000,48000,70000,72000],
     'experience':[3,4,2,3,5,6]}
df=pd.DataFrame(data)
fig= px.bar(df,
            x='Department', 
            y='Salary', 
            title='Department Salary',
            width=800,
            height=500,
            color='Gender',
            hover_data=['experience'],
            pattern_shape='Gender',
            barmode='group',
            text_auto=True,
            orientation='v',
            color_discrete_map={'Male':'yellow','Female':'pink'})
fig.show()
```

# Scatter Plot

```python
import pandas as pd
import plotly.express as px
data={'Department':['IT','IT','HR','HR','Finance','Finance'],
     'Gender':['Male','Female','Male','Female','Male','Female'],
     'Salary':[50000,55000,45000,48000,70000,72000],
     'Hours':[2,3,4,5,6,7]}
df=pd.DataFrame(data)
fig= px.scatter(df,x='Hours', y='Salary', color="Gender")
fig.show()
```

```python
import pandas as pd
import plotly.express as px
data={'Department':['IT','IT','HR','HR','Finance','Finance'],
     'Gender':['Male','Female','Male','Female','Male','Female'],
     'Salary':[50000,55000,45000,48000,70000,72000],
     'Hours':[2,3,4,5,6,7]}
df=pd.DataFrame(data)
fig= px.scatter(df,x='Hours', y='Salary', color="Gender", title='Bubble Scatter Plot', size='Salary', symbol= 'Gender')
fig.show()
```

# 3-D scatter plot

```python
data={'Hours':[2,3,4,5,6],
     'Marks':[40,50,60,70,80],
     'Attendance':[60,65,70,80,90]}
df=pd.DataFrame(data)
fig= px.scatter_3d(df, x='Hours', y='Marks', z='Attendance', color='Marks', size='Attendance')
fig.show()
```

# scatter_matrix

```python
fig= px.scatter_matrix(df, dimensions=['Hours','Marks','Attendance'],color='Marks',title='scatter matrix')
fig.show()
```

```python
data={'Hours':[2,3,4,5,6,7,8,3,5],
     'Marks':[40,50,55,65,70,80,85,48,68],
     'Attendance':[60,70,75,80,85,90,95,72,88],
     'Gender':['Male','Female','Male','Female','Male','Female',
              'Male','Female','Male'],
     'Department':['IT','IT','HR','HR','Finance','Finance',
                  'IT','HR','Finance']}
df=pd.DataFrame(data)
fig=px.scatter(df,x='Hours', y='Marks', 
               width=900, height=500, 
               color='Gender', size='Attendance', 
               color_discrete_map={'Male':'yellow','Female':'pink'}, 
               symbol='Gender', facet_col='Department', facet_row='Gender')
fig.show()
```

# Line chart

```python
data={'Hours':[2,3,4,5,6,7,8,3,5],
     'Marks':[40,50,55,65,70,80,85,48,68]}
df=pd.DataFrame(data)
fig=px.line(df,x='Hours',y='Marks', markers=True, title='Line plot')
fig.show()
```

# Pie Chart

```python
import pandas as pd
import plotly.express as px
data={'Department':['IT','HR','Finance','Marketing'],
     'Employees':[40,20,25,15]}
df=pd.DataFrame(data)
fig=px.pie(df,values='Employees',names='Department',
          color_discrete_sequence=['blue','green','orange','red'],
          title='Employee Distribution by Department')
fig.update_traces(
    textinfo='percent+label',textposition='inside',pull=[0,0.1,0,0])
fig.show()
```

# Histogram

```python
data={'Marks':[40,45,50,55,60,60,65,70,75,80,85],
     'Gender':['Male','Female','Male','Female','Male','Female',
              'Male','Female','Male','Female','Male']}
df=pd.DataFrame(data)
fig=px.histogram(df,x='Marks',color='Gender',
                histfunc='count',pattern_shape='Gender',
                text_auto=True,nbins=10,
                color_discrete_map={'Male':'blue','Female':'pink'})
fig.show()
```

```python
# TASK
import pandas as pd
df=pd.read_csv('production_analysis_dataset.csv')
df
```

```python
# 1 find the total units produced by each product_category.
df.groupby("product_category")['units_produced'].sum()
```

```python
# 2 Calculate the average profit for each factory.
df.groupby("factory")['profit'].mean()
```

```python
# 3 Find the total defective_units for each shift.
df.groupby("shift")['defective_units'].sum()
```

```python
# 4 Calculate the average production_time_hr for each supervisor.
df.groupby("supervisor")['production_time_hr'].mean()
```

```python
# 5 Find the total energy_consumption_kwh for each factory and shift.
df.groupby(['factory','shift'])['energy_consumption_kwh'].mean()
```

```python
# 6  For each product_category, calculate:
# total units_produced
# average profit
# maximum defective_units.
df.groupby('product_category').agg({'units_produced':'sum', 'profit':'mean','defective_units':'max'})
```
