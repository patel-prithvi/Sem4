# Heatmap

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data={"Math":[80,70,60],
     'Science':[75,65,55],
     'English':[90,85,70]}
df=pd.DataFrame(data,index=["Student A",'Student B','Student C'])

sns.heatmap(df,vmax=50,vmin=100)
plt.title('Student Marks Heatmap')
plt.show()
```

# Networkx

```python
import networkx as nx
import matplotlib.pyplot as plt
plt.figure(figsize=(6,6))
G=nx.Graph()
G.add_node(1)
G.add_nodes_from([2,3])
G.add_nodes_from(range(4,7))
G.add_edge(1,2)
G.add_edges_from([(2,3),(3,4),(4,5),(5,6),(6,1)])

nx.draw_networkx(G,node_size=1000,node_color='red',
                width=2,edge_color='blue')
plt.show()
```

# Directed Graph

```python
import networkx as nx
import matplotlib.pyplot as plt
plt.figure(figsize=(6,6))
G=nx.DiGraph()
G.add_node(1)
G.add_nodes_from([2,3])
G.add_nodes_from(range(4,7))
G.add_edge(1,2)
G.add_edges_from([(2,3),(3,4),(4,5),(5,6),(6,1)])

nx.draw_networkx(G,node_size=1000,node_color='red',
                width=2,edge_color='blue')
plt.show()
```

# PB 139

```python
instagram = {'person1': [0,1,1,0,1], 'person2': [0,0,1,0,1], 
              'person3': [1,1,0,1,1],'person4': [1,1,1,0,0],
              'person5': [1,1,0,0,0]}

plt.figure(figsize=(6,6))
G=nx.DiGraph()
Node=list(instagram.keys())
G.add_nodes_from(Node)
Follow=list(instagram.values())

for i in range(len(instagram)):
    for j in range(len(instagram)):
        if Follow[i][j]==1:
            G.add_edge(Node[i],Node[j])

nx.draw_networkx(G,node_size=2200,node_color='red',
                width=2,edge_color='blue')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data={'Department':['IT','IT','IT','HR','HR','HR','Finance',
                   'Finance','Finance'],
     'Gender':['Male','Female','Male','Female','Male','Female',
              'Male','Female','Male'],
     'Salary':[50000,55000,60000,45000,47000,48000,70000,72000,
              71000]}

df=pd.DataFrame(data)

sns.boxplot(x='Department',y='Salary',hue='Gender',data=df,
           color='skyblue')
plt.show()
```

# Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data={'Department':['IT','IT','IT','HR','HR','HR','Finance',
                   'Finance','Finance'],
     'Gender':['Male','Female','Male','Female','Male','Female',
              'Male','Female','Male'],
     'Salary':[50000,55000,60000,45000,47000,48000,70000,72000,
              71000],
     'Ex':[2,3,4,5,6,7,8,9,10]}

df=pd.DataFrame(data)
sns.scatterplot(x='Ex',y='Salary',hue='Gender',style='Gender',
               s=150,data=df)
plt.show()
```
